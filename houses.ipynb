#### 2.1.2. Высота потолков

Начинаем разбираться с данными. Первый по порядку столбец с пропусками ceiling_height - высота потолков. Количественный тип данных. Для заполнения должна подойти среднее или медиана, конечно, если в данных нет большого количества вбросов или аномалий.

```python
data['ceiling_height'].describe()
```

Максимальное значение 100 метров! Наверное, это какая-то труба. Посмотрим на диаграмму рассеяния.

```python
data.plot(x='ceiling_height', y='total_area', kind='scatter')
```

Он один такой, на целом свете. Посмотрим поближе.

```python
data[data['ceiling_height'] == 100]
```

Сто метров высотой и площадью 25 м² за 15 млн. Возможно это какая-то вышка. Сложно сказать реальный это объект или аномалия в данных. Однозначно что он доставит нам хлопот увеличив диапазон значений. В любом случае если от него избавиться, то много не убудет. Так и поступим.

```python
data.drop([22869], inplace = True)
```

Видно что аномальные значения есть, но их немного и среднее попадает. С аномалиями будем разбираться позже, сейчас заполним через группировку по общей площади. Для этого используем функцию 'transform' которой передадим параметр 'mean'. Эта функция создаёт новый столбец, и заполняет его значениями медианы по сгруппированному списку площади квартир. По сути `.transform()` создаёт новый столбец с данными, а после заполняет значениями из столбца пропуски в столбце к которому её применяем.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data.groupby('total_area')['ceiling_height'].transform("mean"))
print('Количество пропусков в столбце "ceiling_height" -', data['ceiling_height'].isna().sum()) #проверим количество пропусков в столбце
```

Оставшиеся пропуски не попавшие в группировку заполним средним.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].mean())
data['ceiling_height'] = round(data['ceiling_height'], 1)
print('Количество пропусков в столбце "ceiling_height" -', data['ceiling_height'].isna().sum())
```

#### 2.1.3. Всего этажей в доме

Следующий столбец floors_total - всего этажей в доме. Количественный тип данных, ведь их можно посчитать.

```python
data['floors_total'].describe()
```

Для заполнения можно использовать медиану c группировкой, аналогично как в предыдущем пункте.

```python
data['floors_total'] = data['floors_total'].fillna(data.groupby('total_area')['floors_total'].transform("mean"))
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].mean())
print('Количество пропусков в столбце "floors_total" -', data['floors_total'].isna().sum())
```

После заполнения нужно проверить не получилось ли так что этаж, где расположена квартира выходит за пределы значений этажей в доме.

```python
data[data['floor'] > data['floors_total']]['floor'].count()
```

Мы наделали аномалий. Хорошо, что немного. Приравняем вышедшие за пределы диапазона значения к последнему этажу здания.

```python
def floor_normalize(row):
    if row[8] > row[6]:
        return row[6]
    else:
        return row[8]
data['floor'] = data.apply(floor_normalize, axis=1)

data[data['floor'] > data['floors_total']]['floor'].count()
```

```python
data
```

Поправили значения в пару строк, уверен что можно короче.

#### 2.1.4. Жилая площадь в квадратных метрах (м²)

Следующее пропущенное значение `living_area`. Заполнение медианой в данном случае не подходит, т.к. для некоторых значений медианное может оказаться больше площади. Чтобы заполнить пропуски возьмём медианное отношение жилой площади к полной площади квартиры. Полученый коэффициент домножим на общую площадь и заполним полученным значением пропуски в жилой площади.

```python
living_area_ratio = data['living_area'].median() / data['total_area'].median() # коэфициент жилой площади
data['living_area'].fillna(data['total_area'] * living_area_ratio, inplace=True)
data['living_area'] = round(data['living_area'], 2)
print('Количество пропусков в столбце "living_area" -', data['living_area'].isna().sum())
```

#### 2.1.5. Апартаменты

Дальше идёт значение is_apartment - апартаменты. Булев тип данных указывающий является ли квартира апартаментами. Установить это по имеющимся данным невозможно. Наиболее подходящий вариант заменить пропуски на 0.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
print('Количество пропусков в столбце "is_apartment" -', data['is_apartment'].isna().sum())
```

#### 2.1.6. Площадь кухни в квадратных метрах (м²)

Дальше идёт столбец `kitchen_area` - площадь кухни в квадратных метрах. Пропуски в этом столбце аналогичны пропускам в `living_area`. Для нахождения коэффициент разделим медиану площади кухни на медиану общей площади.

```python
kitchen_ratio = data['kitchen_area'].median() / data['total_area'].median() # доля площади кухни
data['kitchen_area'].fillna(data['total_area'] * kitchen_ratio, inplace=True)
data['kitchen_area'] = round(data['kitchen_area'], 2)
print('Количество пропусков в столбце "kitchen_area" -', data['kitchen_area'].isna().sum())
```

#### 2.1.7. Число балконов

Пропуски в столбце `balcony` - число балконов, восстановить по имеющимся данным неудасться. Скорее всего пропуски в значениях указывают на отсутсвие балконов, наврятли люди не смогли вспомнить сколько у них балконов и решили оставить пробел. Нет так нет. Заполним пропуски `0`.

```python
data['balcony'] = data['balcony'].fillna(0)
print('Количество пропусков в столбце "balcony" -', data['balcony'].isna().sum())
```

#### 2.1.8. Название населённого пункта

Пропуски в столбце `locality_name` - название населённого пункта восстановить невозможно. Пропущено 49 из 23669 значений. Если удалить эти строки, то это не окажет значительного влияния.

```python
data = data.dropna(subset=['locality_name'])
print('Количество пропусков в столбце "locality_name" -', data['locality_name'].isna().sum())
```

Куда больше вопросов может вызвать наименование населённых пунктов. Проверим какие уникальные значения встречаются.

```python
data['locality_name'].unique()
```

Это очень длинный список, но всё не так плохо, исправить можно. Первое что я вижу это разное написание слова "поселок" тянущее за собой группу неявных дубликатов. Начнём с того что сохраним длину этого списка, т.е. количество уникальных значенией в переменную и после приведём все "поселки" к одинаквому написанию.

```python
def dup_replace(name):
    if 'поселок' in name or 'пос' in name:
        return 'поселок ' + name.split()[-1]
    else:
        return name

data['locality_name'] = data['locality_name'].apply(dup_replace)
```

```python
data.insert(8, 'living_area_ratio', round(data['living_area'] / data['total_area'], 2))
data.insert(9, 'kitchen_area_ratio', round(data['kitchen_area'] / data['total_area'], 2))
```

### 4.8 Факторы влияющие на стоимость картиры в Санкт-Петербурге

Теперь посмотрим как изменяется цена по мере приближения к центру Санкт-Петербург. Для начала нужно определить центр.

```python
spb_apartm = data[data['locality_name'] == 'Санкт-Петербург']
spb_apartm = spb_apartm.dropna()
spb_apartm.insert(24, 'city_centers_nearest_km', spb_apartm['city_centers_nearest'] / 1000)
spb_apartm['city_centers_nearest_km'] = spb_apartm['city_centers_nearest_km'].astype('int')
```
